```python
import matplotlib.pyplot as plt
import random
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['Arial Unicode MS', 'sans-serif']
```

```python
# 指定画板的大小等等
plt.figure(figsize=(6, 6), dpi=100)

# 指定axis的一些坐标点,必须是列表
x = [1,2,3,4,5,6,7]
y = [17,17,18,15,11,11,13]

# 画出折线图
plt.plot(x, y)

# 必须在show()的前面调用
plt.savefig("./test.png")

# 最终显示图
plt.show()
```

```python
# 实现温度的画图
plt.figure(figsize=(20, 8), dpi=80)

#指定x分钟,y温度的值
x = range(60)
y = [random.uniform(15, 18) for i in range(60)]

# 添加北京的温度变化
y_beijing = [random.uniform(1, 3) for i in range(60)]

# 通过plot折线图显示,一般放在最前面显示
plt.plot(x, y, label="上海")

# 画出北京的温度
plt.plot(x, y_beijing, label="北京", color='r', linestyle="--")


# 自定义刻度值,指定每个刻度的名称[0, 5, 10, 15....]      [11点0分，11点5分]
x_name = ["11点{}分".format(i) for i in range(0, 60)]
plt.xticks(x[::5], x_name[::5])
# 构造显示y轴的范围
y_ = [i for i in range(0, 50)]
plt.yticks(y_[::5])

# 指定x坐标的名字，y坐标的名字，标题名称
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("11点到12点之间的温度变化")

# 显示数据的描述
plt.legend(loc="best")


plt.show()
```

```python
# 显示不同的电影票房数据对比（条形图）
plt.figure(figsize=(20, 8), dpi=80)

# 准备数据，电影名称，票房
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]

# 画坐标的数据必须是数值型
x = range(len(movie_name))

# 显示条形图
plt.bar(x, y, width=0.2, color=['b','r','g','y','c','m','y','k','c','g','g'])

# 显示电影名字刻度
plt.xticks(x, movie_name)

# 对横坐标纵坐标标题描述
plt.xlabel("电影名称")
plt.ylabel("电影票房")
plt.title("11月中国电影票房")

# 显示
plt.show()
```

```python
# 比较不同电影的相同时间票房对比
plt.figure(figsize=(20, 8), dpi=80)

# 电影名称，首日首周的票房
movie_name = ['雷神3：诸神黄昏','正义联盟','寻梦环游记']
first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
x = range(len(movie_name))

# 显示
plt.bar(x, first_day, width=0.2, label="首日票房")
plt.bar([i+0.2 for i in x], first_weekend, width=0.2, label="首周票房")

# 修改x轴的描述,显示的位置
plt.xticks([i+0.1 for i in x], movie_name)

# 显示数据的描述
plt.legend(loc="best")
plt.xlabel("电影名称")
plt.title("首日首周电影票房对比")

plt.show()
```
